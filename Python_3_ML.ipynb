```python
X = advertising['TV']
y = advertising['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
len(X_test)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
import statsmodels.api as sm
```

```python
X_train_sm = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

```python

print(lr.summary())

#Արդյունքում՝
#1. TV գործակիցը 0,054 է, շատ ցածր p արժեքով - Գործակիցը վիճակագրորեն նշանակալի է։ Այնպես որ, ասոցիացիան զուտ պատահական չէ։
#2. R-squared 0,816 է - Նշանակում է, որ վաճառքի տատանումների 81.6%-ը բացատրվում է հեռուստատեսությամբ
#3. F statistic շատ ցածր p արժեք ունի (գործնականում ցածր) - Նշանակում է, որ մոդելի համապատասխանությունը վիճակագրորեն նշանակալի է, և բացատրված շեղումը զուտ պատահական չէ:

# const 6.9487 - b0-n e, TV 0.0545 - b1-n e
#mer sales-i bazayin arjeqy 6.9487 e, aysinqn aranc gorconneri, mer sales-y misht 6.9487 e
#TV-i govazdi vra 1 miavor caxs avelanalu depqum` mer sales-y misht poxvum e 0.05 miavorov
# P>|t| - ete sa 0.05-ic bardzr e, da vat e. Mer mot 0.05-ic poqr e: P-n ayn mijakayqn e, vory stugum e`ardyoq mer gorconnery vijakagrakanoren nshanakali en te che
  #ete da 0.1 liner /0.05-ic mec/, menq ayd gorcony voch nshanakali khamareinq mer datayi mej

  #0.05 նշանակում է 95տոկոս նշանակալիության միջակայքում է, այսինքն կարող եմ 95տոկոսով ասեմ, որ իմ գործակիցը 0 չի, այսինքն 0-ից տարբերվում է․
  #եթե գծային ռեգրեսիոն մոդելում ինքը 0 լիներ, նշանակում է, որ տվյալ գործոնը ոչ նշանակալի կլիներ։
  #Եթե 0․1 լիներ, նշանակում է, որ ավելի քիչ նշանակալի կլիներ, 90 տոկոս հավանականությամբ կլիներ։ Մոդելը գոնե 95տոկոս նշանակալիություն պիտի ունենա։



#	R-squared: կախյալ և անկախ փոփոխականների միջև կապի մոդելավորումն է,  	0.816 - որը նշանակում է` կախյալ փոփոխականի 81,6%-ը բացատրում է անկախ փոփոխականի վարիացիան:
# Adj. R-squared /Կարգավորված R²/: ցույց է տալիս ռեգրեսիոն մոդելի փոփոխությունը անկախ փոփոխականներից մեկի գործակցի զրո լինելու դեպքում: Եթե որևէ գործակցի զրո լինելու դեպքում մոդելի որակը չի փոխվում, ուրեմն գնահատականը լավ է: Adj. R²-ն հաշվի է առնում տվյալների քանակը։
#F-statistic: վարկածների ստուգման գործիք է:	Ցույց է տալիս, որ մոդելը զգալիորեն ավելի լավ է, քան պատահական կանխատեսումը։
#Prob (F-statistic): F-statistic-ի հավանականությունն է: Կարող է լինել 0; 0,001; 0,01; 0,05:
#coef (TV) -   0.0545 - Յուրաքանչյուր 1 միավոր TV աճը ավելացնում է Sales-ը 0.0545 միավորով
#Durbin-Watson - օգտագործվում է գծային ռեգրեսիայի դեպքում՝ ինքնակապվածություն (autocorrelation) հայտնաբերելու համար։
  #Եթե ռեգրեսիոն մոդելի սխալները (residuals) համակարգված ձևով կապ ունեն միմյանց հետ,ապա ասում ենք, որ տվյալներում կա ինքնակապվածություն
  #Այն տատանվում է [0;4] միջակայքում: Եթե 2-ի շրջակայքում է, ապա ճիշտ են ընտրված արժեքները, եթե մեծ է 2-ից` կա բացասական կորելացիա, և 2-ից փոքր լինելու դեպքում կա դրական կորելացիա
```

```python
#9. Declare feature vector and target variable
#Սահմանենք x և y-ները
X = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
#10. Split data into separate training and test set
# բաժանենք X և y-ը training և testing հավաքածուների

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#ստուգենք բացակայող արժեքները X_train-ում՝
X_train.isnull().sum()
```

```python
# ստուգենք բացակայող արժեքները X_test-ում
X_test.isnull().sum()
```

```python
#օգտագործենք ցիկլը X_train սյունակներում դատարկ արժեքների տոկոսը հայտնաբերելու համար՝
for col in X_train.columns:
    if X_train[col].isnull().mean()>0:
        print(col, round(X_train[col].isnull().mean(),4))
```

```python
#X_train և X_test տվյալների հավաքածուներում լրացնենք դատարկ արժեքները մեդիանայով
for df1 in [X_train, X_test]:
    for col in X_train.columns:
        col_median=X_train[col].median()
        df1[col].fillna(col_median, inplace=True)
```

```python
# կրկին ստուգենք X_train-ի թվային փոփոխականների բացակայող արժեքները
X_train.isnull().sum()
```

```python
#12. Feature Scaling
#Այժմ մենք ունենք ուսուցման և թեստավորման հավաքածու: Իսկ հիմա պիտի քարտեզագրենք բոլոր հատկանիշների փոփոխականները նույն սանդղակի վրա, որը կոչվում է հատկանիշի մասշտաբավորում /feature scaling/:

cols = X_train.columns
```

```python
#իրականացնենք տվյալների ստանդարտացում՝Xnew=(X−μ)/σ
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])
```

```python
X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
X_train.head()
```

```python
#13. Fit K Neighbours Classifier to the training eet

# ներմուծեք KNeighbors ClaSSifier-ը sklearn-իցё
from sklearn.neighbors import KNeighborsClassifier
#կառուցենք K-Nearest Neighbors դասակարգման մոդելը ըստ 3 ամենամոտ հարևանների
knn = KNeighborsClassifier(n_neighbors=3)
#fit() մեթոդը մոդելը հարմարեցնում է ուսումնական հավաքածուին՝ X_train (features) և y_train (target):
knn.fit(X_train, y_train)
```

```python
#14. Predict test-set results
y_pred = knn.predict(X_test)
y_pred
```

```python
#predict_proba մեթոդը
#predict_proba մեթոդը տվյալ դեպքում տալիս է թիրախ փոփոխականի (2 և 4) հավանականությունները զանգվածի տեսքով:
#2-ը բարորակ քաղցկեղի հավանականության համար է, իսկ 4-ը՝ չարորակ քաղցկեղի հավանականության համար:

# արդյունք ստանալու հավանականությունը որպես 2 - բարորակ քաղցկեղ
knn.predict_proba(X_test)[:,0]
```

```python
#արդյունք ստանալու հավանականությունը որպես 4 - չարորակ քաղցկեղ
knn.predict_proba(X_test)[:,1]
```

```python
#15. Check accuracy score
#իրականացնենք մոդելի ճշգրտության (accuracy) գնահատում՝
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))

#Այստեղ y_test-ը ճշմարիտ դասի լեյբլներն են, իսկ y_pred-ը կանխատեսված դասի լեյբլներն են թեստային հավաքածուում:
```

```python
#կանխատեսենք X_train տվյալները և արդյունքը պահենք y_pred_train փոփոխականում
y_pred_train = knn.predict(X_train)
```

```python
#գնահատենք մոդելի ուսումնական հավաքածուի (training set) ճշգրտությունը (accuracy)
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
#Ստուգեք overfitting և underfitting
# տպեք training և test հավաքածուների scor-երը
print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))

#training հավաքածուի ճշգրտության միավորը 0,9821 է, մինչդեռ test հավաքածուի ճշգրտությունը 0,9714 է:
#Այս երկու արժեքները բավականին համադրելի են։ Այնպես որ, գերհարմարվելու մասին խոսք լինել չի կարող։
```

```python
# Փորձարկում ենք K-ի տարբեր արժեքներ (1-ից 20)
k_values = range(1, 21)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_k = knn.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred_k))

# Կառուցում ենք գրաֆիկ՝ K-ի արժեքների և ճշգրտության միջև կապը տեսնելու համար
plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracy_scores, marker='o', linestyle='-', color='b')
plt.xlabel("K-ի արժեքը")
plt.ylabel("Ճշգրտություն (Accuracy)")
plt.title("KNN-ի ճշգրտության կախվածությունը K-ի արժեքից")
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Գտնում ենք լավագույն K արժեքը
best_k = k_values[accuracy_scores.index(max(accuracy_scores))]
best_k, max(accuracy_scores)
```

```python
#Classification accuracy
TP = cm_7[0,0]
TN = cm_7[1,1]
FP = cm_7[0,1]
FN = cm_7[1,0]
```

```python
# տպենք classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))
```

```python
#Classification error
classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
#Precision /Ճշգրտություն/
#Precision-ը կարող է սահմանվել որպես բոլոր կանխատեսված դրական արդյունքներից ճիշտ կանխատեսված դրական արդյունքների տոկոսը:

precision = TP / float(TP + FP)
print('Precision : {0:0.4f}'.format(precision))
```

```python
#Recall
#Recall-ը կարող է սահմանվել որպես բոլոր փաստացի դրական արդյունքներից ճիշտ կանխատեսված դրական արդյունքների տոկոսը:
#Recall-ը բացահայտում է ճիշտ կանխատեսված իրական դրականների համամասնությունը:
recall = TP / float(TP + FN)
print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

```python
#True Positive Rate
#սա հոմանիշ է Recall-ին
true_positive_rate = TP / float(TP + FN)
print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
#False Positive Rate
false_positive_rate = FP / float(FP + TN)

print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
#Specificity
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```

```python
#f1-score
#f1-score=2*(precision*recall)/(precision+recall): Ամենալավ հնարավոր f1 միավորը կլինի 1.0, իսկ վատագույնը՝ 0.0:
```

```python
#19. ROC-AUC
#Դասակարգման մոդելի կատարողականությունը տեսողականորեն չափելու մեկ այլ գործիք ROC Curve-ն է: ROC Curve-ը նշանակում է ընդունիչի գործառնական բնութագրիչ կոր: ROC կորը գծապատկեր է, որը ցույց է տալիս դասակարգման մոդելի կատարումը տարբեր դասակարգման շեմերի մակարդակներում:
#ROC կորը գծագրում է True Positive Rate (TPR) ընդդեմ False Positive Rate (FPR) տարբեր շեմային մակարդակներում:
#True Positive Rate (TPR) կոչվում է նաև հետկանչում: Այն սահմանվում է որպես TP-ի հարաբերակցությունը (TP + FN):
#False Positive Rate (FPR) սահմանվում է որպես FP-ի հարաբերակցությունը (FP + TN):
#ROC կորի մեջ մենք կկենտրոնանանք մեկ կետի TPR-ի (True Positive Rate) և FPR-ի (False Positive Rate) վրա: Սա մեզ կտա ROC կորի ընդհանուր կատարումը, որը բաղկացած է TPR-ից և FPR-ից տարբեր շեմային մակարդակներում:
#Այսպիսով, ROC կորը գծագրում է TPR-ն ընդդեմ FPR-ի տարբեր դասակարգման շեմերի մակարդակներում: Եթե մենք իջեցնենք շեմային մակարդակները, դա կարող է հանգեցնել նրան, որ ավելի շատ տարրեր դասակարգվեն որպես դրական: Դա կավելացնի ինչպես True Positives (TP), այնպես էլ False Positives (FP):
```

```python
# plot ROC Curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_1, pos_label=4)
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0,1], [0,1], 'k--' )
plt.rcParams['font.size'] = 12
plt.title('ROC curve for Breast Cancer kNN classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')

plt.show()

#ROC կորը մեզ օգնում է ընտրել շեմային մակարդակ, որը հավասարակշռում է զգայունությունն /sensitivity / ու առանձնահատկությունը /specificity / որոշակի համատեքստի համար:
```

```python
#ROC AUC
#ROC AUC-ն նշանակում է ընդունիչի գործառնական բնութագիր /Receiver Operating Characteristic/՝ կորի տակ գտնվող տարածք /Area Under Curve/:
#Դա դասակարգչի կատարողականը համեմատելու տեխնիկա է: Այս տեխնիկայում մենք չափում ենք կորի տակ գտնվող տարածքը (AUC): Կատարյալ դասակարգիչը կունենա ROC AUC հավասար 1-ի, մինչդեռ զուտ պատահական դասակարգիչը կունենա ROC AUC հավասար 0,5:
#Այսպիսով, ROC AUC-ն ROC plot-ի տոկոսն է, որը գտնվում է կորի տակ:

# հաշվարկենք ROC AUC
from sklearn.metrics import roc_auc_score
ROC_AUC = roc_auc_score(y_test, y_pred_1)
print('ROC AUC : {:.4f}'.format(ROC_AUC))

#ROC AUC-ն դասակարգչի կատարողականի մեկ թվային ամփոփագիր է: Որքան մեծ է արժեքը, այնքան լավ է դասակարգիչը:
#Մեր մոդելի ROC AUC-ը մոտենում է 1-ին: Այսպիսով, մենք կարող ենք եզրակացնել, որ մեր դասակարգիչը լավ աշխատանք է կատարում կանխագուշակելու համար՝ դա բարորակ կամ չարորակ քաղցկեղ է:
```

#ML - Decision tree /bank client data/

The classification goal is to predict if the client will subscribe (yes/no) a term deposit (variable y).

Attribute information

bank client data:

* age (numeric)
*job : type of job (categorical: 'admin.','blue-collar','entrepreneur','housemaid','management','retired','self-employed','services','student','technician','unemployed','unknown')
*marital : marital status (categorical: 'divorced','married','single','unknown'; note: 'divorced' means divorced or widowed)
*education (categorical: 'basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course','university.degree','unknown')
*default: has credit in default? (categorical: 'no','yes','unknown')
*housing: has housing loan? (categorical: 'no','yes','unknown')
*loan: has personal loan? (categorical: 'no','yes','unknown')
*contact: contact communication type (categorical: 'cellular','telephone')
*month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')
*day_of_week: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')
*duration: last contact duration, in seconds (numeric). Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.
other attributes:

* campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
* pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
* previous: number of contacts performed before this campaign and for this client (numeric)
* poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')

social and economic context attributes

* emp.var.rate: employment variation rate - quarterly indicator (numeric)
* cons.price.idx: consumer price index - monthly indicator (numeric)
* cons.conf.idx: consumer confidence index - monthly indicator (numeric)
* euribor3m: euribor 3 month rate - daily indicator (numeric)
* nr.employed: number of employees - quarterly indicator (numeric)
*Output variable (desired target):

y - has the client subscribed a term deposit? (binary: 'yes','no')
